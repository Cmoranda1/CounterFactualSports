```python
#How important is it to find a franchise QB?
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as sm

num_qbs_vs_win_pct = pd.read_csv("66_75_qb_vs_wpct.csv")
num_qbs_vs_win_pct['wpct2'] = num_qbs_vs_win_pct['wpct'] * 100
fig1 = sns.lmplot(x='qbnum', y='wpct2',  data=num_qbs_vs_win_pct)
plt.xlabel('Number of Primary QBs')
plt.ylabel('Winning Percentage')
plt.title("Relationship Between Number of Primary Qbs and Team Win Pct between '66-'75", fontsize=20)
fig1.savefig("regression1.png")
```

```python
reg1 = sm.ols(formula = 'wpct ~ qbnum', data= num_qbs_vs_win_pct).fit()
print(reg1.summary())
```
