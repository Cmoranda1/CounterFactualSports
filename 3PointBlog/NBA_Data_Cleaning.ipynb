```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime

NaN = np.nan

shot = pd.read_csv("Data/shot_logs_16_17.csv")
games = pd.read_csv("Data/games.csv")
teams = pd.read_csv("Data/nba_teams.csv")
games = games[(games['GAME_DATE_EST'] >= "2016-10-25") & (games['GAME_DATE_EST'] <= "2017-04-12")]
games['HOME_TEAM_ABBR'] = ''
games['AWAY_TEAM_ABBR'] = ''
games['HOME_X_MAKES'] = NaN
games['HOME_Y_MAKES'] = NaN
games['AWAY_X_MAKES'] = NaN
games['AWAY_Y_MAKES'] = NaN

games['HOME_X_MAKES'] = games['HOME_X_MAKES'].astype(object)
games['HOME_Y_MAKES'] = games['HOME_Y_MAKES'].astype(object)
games['AWAY_X_MAKES'] = games['AWAY_X_MAKES'].astype(object)
games['AWAY_Y_MAKES'] = games['AWAY_Y_MAKES'].astype(object)

shot = shot[shot['current_shot_outcome'] == 'SCORED']
shot['date'] = pd.to_datetime(shot['date'])
games['GAME_DATE_EST'] = pd.to_datetime(games['GAME_DATE_EST'])
shot.head()
```

```python
for index, row in games.iterrows():
    games.at[index, 'HOME_TEAM_ABBR'] = teams[teams['TEAM_ID'] == row['HOME_TEAM_ID']].ABBREVIATION.values[0]
    games.at[index,'AWAY_TEAM_ABBR'] = teams[teams['TEAM_ID'] == row['VISITOR_TEAM_ID']].ABBREVIATION.values[0]
games
```

```python
for index, row in games.iterrows():
    shot_game = shot[(shot['home_team'] == row['HOME_TEAM_ABBR']) & (shot['away_team'] == row['AWAY_TEAM_ABBR']) & (shot['date'] == row['GAME_DATE_EST'])]
    home_x_makes = list(shot_game[shot_game.home_game =='Yes'].location_x)
    home_y_makes = list(shot_game[shot_game.home_game =='Yes'].location_y)
    
    away_x_makes = list(shot_game[shot_game.home_game =='No'].location_x)
    away_y_makes = list(shot_game[shot_game.home_game =='No'].location_y)
    
    games.at[index, 'HOME_X_MAKES'] = home_x_makes
    games.at[index, 'HOME_Y_MAKES'] = home_y_makes
    
    games.at[index, 'AWAY_X_MAKES'] = away_x_makes
    games.at[index, 'AWAY_Y_MAKES'] = away_y_makes
```

```python
games.to_csv('../Desktop/shot_data_by_game.csv')
```
