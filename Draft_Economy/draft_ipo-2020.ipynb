```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ipo_21 = pd.read_csv("ipos-2020.csv")
```

```python
ipo_21 = ipo_21.sort_values("IPO Price", ascending=False)
```

```python
ipo_21_top256 = ipo_21.head(225)
ipo_21_top256["Round"] = 0
ind = 0
roundnum = 0
picks = []
rounds = []
for i in ipo_21_top256.iterrows():
    if(ind % 32 ==0):
        roundnum += 1
    rounds.append(roundnum)
    ind += 1
    picks.append(ind)
ipo_21_top256["Round"] = rounds
ipo_21_top256["Pick"] = picks
print(picks)
```

```python
ipos_positive_return = ipo_21_top256[ipo_21_top256["Return"]>0]
```

```python
ipos_negative_return = ipo_21_top256[ipo_21_top256["Return"]<0]
```

```python
draft_21 = pd.read_csv("2020_draft.csv")
```

```python
draft_21["wAV"]
```

```python
draft_value = pd.read_csv("value_chart.csv")
```

```python
draft_value.drop(["Year", "Tm"], axis = 1)
```

```python
draft_value = draft_value.head(225)
draft_value["Multiplier"] = draft_value["Value"]/3000
```

```python
draft_21 = draft_21.merge(draft_value, how='left', on='Pick')
```

```python
draft_21["AdjustedValue"] = draft_21["wAV"] * draft_21["Multiplier"]
```

```python
draft_21["AvgValue"] = 21.6 * draft_21["Multiplier"]
```

```python
draft_21["AVoverAvg"] = draft_21["wAV"] - draft_21["AvgValue"]
```

```python
draft_21 = draft_21.head(225)
```

```python
performances = plt.scatter(draft_21.Pick, draft_21.wAV)
average = plt.plot(draft_21.Pick, draft_21.AvgValue, color='red')
```

```python
draft_21["Return"] = draft_21["AVoverAvg"]/draft_21["AvgValue"]
```

```python
draft_21 = draft_21.merge(ipo_21_top256, how='left', on='Pick')
```

```python
draft_21["IPOMultiplier"] = draft_21["IPO Price"]/120
draft_21["AdjustedValueIPO"] = draft_21["wAV"] * draft_21["IPOMultiplier"]
draft_21["AvgValueIPO"] = 21.6 * draft_21["IPOMultiplier"]
draft_21["AVoverAvgIPO"] = draft_21["wAV"] - draft_21["AvgValueIPO"]
draft_21["ReturnIPO"] = draft_21["AVoverAvgIPO"]/draft_21["AvgValueIPO"]
performances = plt.scatter(draft_21.Pick, draft_21.wAV)
average = plt.plot(draft_21.Pick, draft_21.AvgValueIPO, color='red')
```
